```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading csv file (Dataset)
wine = pd.read_csv('winequality-red.csv')
```

```python
# Displaying first 10 rows of dataset
wine.head(10)
```

```python
# Creating catagories based on Quality Ratings
def quality(wine):
    if (wine['Quality Rating'] < 5):
        return "Poor"
    elif (wine['Quality Rating'] < 8):
        return "Average"
    else:
        return "Excellent"
```

```python
# Adding a new column 'Quality'
wine['Quality'] = wine.apply(quality, axis=1)
```

```python
# Displaying first 10 rows of updated dataset
wine.head(10)
```

```python
# Displaying no. of rows and columns of dataset respectively
wine.shape
```

```python
# Checking for missing values
missingValues = wine.isna().sum()
missingValues
```

```python
# Displaying no. of wine samples of each quality
wine['Quality'].value_counts()
```

```python
# Displying basic statistical details like count, mean,standard deviation, minimum value, maximum value etc. of the data 
wine.describe()
```

```python
# Plotting histogram for Quality Rating
plt.hist(wine['Quality Rating'], bins = 6)
plt.title('Distribution of the Quality Rating')
plt.xlabel('Quality Ratings')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting Cluster Map
sns.clustermap(wine.corr(), annot=True, linewidth=0.1, cmap = 'Reds')
```

```python
# Jointplot: Density vs Fixed Acidity
sns.jointplot(y='Density',x='Fixed Acidity',data=wine,kind='hex')
```

```python
# Jointplot: pH vs Citric Acid
sns.jointplot(x='pH',y='Citric Acid',data=wine,kind='kde')
```

```python
# Jointplot: Total Sulfur Dioxide vs Free Sulfur Dioxide
sns.jointplot(x='Total Sulfur Dioxide', y='Free Sulfur Dioxide',data=wine, kind='reg')
```

```python
# Jointplot: Citric Acid vs Fixed Acidity
sns.jointplot(x='Citric Acid', y='Fixed Acidity',data=wine, kind='reg')
```

```python
# Jointplot: Volatile Acidity vs Citric Acid
sns.jointplot(x='Volatile Acidity', y='Citric Acid',data=wine, kind='reg')
```

```python
# Jointplot: pH vs Fixed Acidity
sns.jointplot(x='pH', y='Fixed Acidity',data=wine, kind='hex')
```

```python
# Barplot: Quality Rating vs Sulphates
sns.barplot(x = 'Quality Rating', y='Sulphates', data = wine)
```

```python
sns.set_style('whitegrid')
sns.set_context('poster',font_scale=1)

properties = ["Fixed Acidity","Volatile Acidity","Citric Acid","Residual Sugar","Chlorides","Free Sulfur Dioxide","Total Sulfur Dioxide","Density","pH","Sulphates","Alcohol"]
wines = wine[properties]
data = pd.DataFrame(wines, columns = wines.columns.values.tolist())
data[:3].T.plot(kind='bar',figsize=(22,10), color=['red','yellow','green'])
plt.title('Comparison of Chemical Properties by Quality')
plt.legend(('Poor','Average','Excellent'))
plt.xlabel('Chemical Properties')
plt.ylabel('Value')
plt.show()
```
